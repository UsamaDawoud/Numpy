```python
# in image processing a lot of things can done 
# example --- first import some library for disply to make the operation cleare
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import cv2
```

```python
def load_img():
    blank_img = np.zeros((600,600))
    font = cv2.FONT_HERSHEY_SIMPLEX
    cv2.putText(blank_img , org = ( 50 , 300 ) , text ='ABCDE' , color = (255,255,255),fontFace=font ,
                fontScale = 5 ,thickness = 25)
    return blank_img
```

```python
def display_img(img):
    fig=plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap='gray');
```

```python
img = load_img()
```

```python
display_img(img)
```

```python
# to add noise for it 
white_noise= np.random.randint(low=0 , high=2 , size=(600,600)) * 255 
display_img(white_noise)
```

```python
# here it is possible to make ( add operation) for this two images !! like :
result = img + white_noise
display_img(result)
```

```python
# or in aother scenario 
black_noise = np.random.randint(low=0 , high=2 , size=(600,600)) * -255
bnoise_img = img + black_noise
bnoise_img[bnoise_img == -255 ] = 0
display_img(bnoise_img)
```
